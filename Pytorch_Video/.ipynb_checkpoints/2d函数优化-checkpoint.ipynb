Himmelblau function

用来检测优化器是否有很好的解

```python
import torch
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
```

```python
import numpy as np
```

```python
def himmelblau(x):
    return (x[0] ** 2 + x[1] - 11) ** 2 + (x[0] + x[1] ** 2 - 7) ** 2
```

```python
x = np.arange(-6, 6, 0.1)
y = np.arange(-6, 6, 0.1)
```

```python
print('x,y range:',x.shape, y.shape)
```

```python
X, Y = np.meshgrid(x,y) # 生成坐标
```

```python
print("X,Y maps:",X.shape, Y.shape)
```

```python
Z = himmelblau([X, Y])
```

```python
fig = plt.figure("himmelblau")
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, Z)
ax.view_init(60, -30)
ax.set_label('x')
ax.set_label('y')
plt.show()
```

```python
x = torch.tensor([0.,0.], requires_grad=True)  # x初始化为[0,0]，并设置为需求导
```

```python
optimizer = torch.optim.Adam([x], lr = 1e-3)  # learning rate
```

```python
for step in range(20000):
    pred = himmelblau(x)
    optimizer.zero_grad()   # 将梯度信息清零
    pred.backward()   # 获取当前x与y的梯度信息
    optimizer.step()    # 更新x与y的值
    if step % 2000 == 0:
        print('Step {}: x = {}, f(x) = {}'.format(step, x.tolist(), pred.item()))
```
